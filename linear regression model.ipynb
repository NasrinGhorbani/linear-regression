```python
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook

ws=load_workbook('linear_regression_data.xlsx')['Sheet1']
max_row=ws.max_row

data=[]
for i in range(1, max_row+1):
	data.append([ws[f"A{i}"].value, ws[f"B{i}"].value, ws[f"C{i}"].value, ws[f"D{i}"].value])

data.pop(0)
data=np.array(data)

x=data[:,0:3].reshape((-1,3))
y=data[:, 3].reshape((-1,1))
```

```python
print(x, x.shape)
print(y, y.shape)
```

```python
print(data)
```

```python
plt.plot(x[:,0],y,'o', label='input1')
plt.plot(x[:,1],y,'o', label='input2')
plt.plot(x[:,2],y,'o', label='input3')
plt.legend()
plt.title('Input - Real Output')
plt.xlabel('input')
plt.ylabel('output')
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
x_smp=[[730,720,760]]
x_smp=np.array(x_smp)
y_smp=model.predict(x_smp)
print(y_smp)
```

```python
y_hat=model.predict(x)
plt.plot(x[:,0],y_hat,'o', label='input1')
plt.plot(x[:,1],y_hat,'o', label='input2')
plt.plot(x[:,2],y_hat,'o', label='input3')
plt.legend()
plt.title('Input - Predicted Output')
plt.xlabel('input')
plt.ylabel('predicted output')
```

```python
plt.plot(y,y_hat,'o')
plt.title('Real Output - Predicted Output')
plt.xlabel('Real Output')
plt.ylabel('Predicted Output')
```
